```python
#Importing the python libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
```

```python
#Loading the dataset
euro_energy = pd.read_csv("EuroEnergy.csv")
```

```python
#Displaying the dataset
euro_energy
```

```python
#Renaming the columns
euro_energy.rename(columns={"Unnamed: 0":"Country","gdp":"Gross Domestic Product(Million 1975 USD)","energy":"Energy Consumption(Million Kg Coal Equivalence)"}, inplace=True)
```

```python
euro_energy
```

```python
#Displaying a statistical summary of the dataset
euro_energy.describe()
```

```python
#Exploring the relationship between GDP and Energy Consumption using scatter plot
sns.scatterplot(data=euro_energy, x="Gross Domestic Product(Million 1975 USD)",y="Energy Consumption(Million Kg Coal Equivalence)")
```

```python
#Representing the dataset set in a simple bar chart
euro_energy.plot(x="Country", kind="barh")
```

```python
top_five_countries = euro_energy.nlargest(5,"Gross Domestic Product(Million 1975 USD)")
```

```python
#Showing the top 5 countries with the highest GDP
sns.barplot(data=top_five_countries, x="Country", y="Gross Domestic Product(Million 1975 USD)")
```

```python
sns.scatterplot(data=euro_energy, x="Gross Domestic Product(Million 1975 USD)",y="Energy Consumption(Million Kg Coal Equivalence)",hue=("Country"))
```
